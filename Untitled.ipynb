```python
import numpy as np
import pandas as pd
import math
from scipy import stats
```

```python
df = pd.read_csv('data/background_color_experiment.csv')
df
```

```python
control_group = df[df['user_type']=='control']['session_duration']  #here control group represents the group who still used the original or old website
variation_group = df[df['user_type']=='variation']['session_duration']  #here variation means the group who uses the new website
```

```python
mean_control = control_group.mean()  
mean_variation = variation_group.mean()
print(f'The average session duration of old website is: {mean_control:.2f}')
print(f'The average session duration of new website is: {mean_variation:.2f}')
```

```python
#It seems the new website is working as it increased the average session duration of the website
#but to confirm this, we must prove this with hypothesis testing
#converting into numpy array
X_c = control_group.to_numpy()
X_v = variation_group.to_numpy()
X_c
```

```python
def getstats(X):
    n = len(X)  #number of sample datas
    x = np.mean(X)
    s = np.std(X,ddof=1)  #passing the degree of freedom as 1
    return (n,x,s)
n_c,mean_c,s_c = getstats(X_c)
n_v,mean_v,s_v = getstats(X_v)
print(f'Stats of session duration of old website, mean is: {mean_c} and std is: {s_c}')
print(f'Stats of session duration of new website, mean is: {mean_v} and std is: {s_v}')

    
```

```python
def degrees_of_freedom(n_v, s_v, n_c, s_c):
   
    s_v_n_v = np.square(s_v) / n_v
    s_c_n_c =  np.square(s_c) / n_c
    numerator = np.square(s_v_n_v + s_c_n_c)
    denominator = np.square(s_v_n_v)/(n_v - 1) + np.square(s_c_n_c)/(n_c - 1)
    dof = numerator/denominator
        
    return dof
d = degrees_of_freedom(n_v, s_v, n_c, s_c)
print(f"The degrees of freedom for the t-student in this scenario is: {d:.2f}")    
```

```python
def t_value(x_v,x_c,s_v,s_c,n_v,n_c):
    s_v_n_v = np.square(s_v) / n_v
    s_c_n_c = np.square(s_c) / n_c
    numerator = x_v - x_c
    denominator = np.sqrt(s_v_n_v + s_c_n_c)
    t = numerator / denominator
    return t
t = t_value(mean_v,mean_c,s_v,s_c,n_v,n_c)
print(f"The t-value for this experiment is: {t:.2f}")    
```

```python
#calculation of p value
def p_value(d,t_value):
    t_d = stats.t(df= d)   #getting the t_distribution based on degree of freedom
    pvalue = 1 -t_d.cdf(t_value)  #calculating p value based on area to the left or area under the t distribution
    return pvalue
print(f"The p-value for t_15 with t-value = 1.10 is: {p_value(15, 1.10):.4f}")
print(f"The p-value for t_30 with t-value = 1.10 is: {p_value(30, 1.10):.4f}")    
```
